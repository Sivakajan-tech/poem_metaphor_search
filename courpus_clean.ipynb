```python
import pandas as pd

df = pd.read_csv('data/corpus.csv')
df.head()
```

```python
import csv
import json

# Open the CSV file in read mode with encoding='utf-8'
with open("data/corpus.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.reader(csv_file)

    # Create a list to hold the CSV data
    data = []

    # Iterate over the CSV rows and store them in the data list
    for row in csv_reader:
        data.append(row)

# Create a new JSON file in write mode with encoding='utf-8'
with open("data/corpus.json", "w", encoding="utf-8") as json_file:
    # Convert the CSV data to a list of dictionaries
    keys = data[0]  # Assuming the first row contains column headers
    json_data = [dict(zip(keys, row)) for row in data[1:]]

    # Write the JSON data to the JSON file
    json.dump(json_data, json_file, ensure_ascii=False, indent=4)

# No need to explicitly close the files; 'with' handles that for you
```
